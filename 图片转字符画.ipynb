```python
from PIL import Image
```

```python
im = Image.open('QQ截图20181219124204.png')
```

```python
ascii_char = list(r"$kB%8&WM#*oa&**@qwmZO0QLCJUYXzcvun<rjft/\|()1{}[]?-_+~<>i!lI;:,\ ^`'. ")
```

```python
ascii_char = list(r"爱你龙洋小宝贝。")
```

```python
im = im.resize(tuple((int((im.size[0])/1),int((im.size[1])/1.5))))
im.size
```

```python
text = ''

for h in range(im.size[1]):
    for w in range(im.size[0]):
        gray = int((19595 * im.getpixel((w, h))[0] + 38469 * im.getpixel((w, h))[1] + 7474 * im.getpixel((w, h))[2]) >> 16)
        unit = 256.0 / len(ascii_char)
        text += ascii_char[int(gray / unit)]
    text += '\n'
```
